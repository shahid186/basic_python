```python
print("mcq")

q={}
a=[]
ans={}
for i in range(10):
    a=[]
    inp=input("enter {} question ".format(i+1))
    for j in range (4):
        a.append(input("give{} options  ".format(j+1)))
    q[inp]=a
    ans[inp]=(input("give the correct option [a,b,c,d]:"))
print(q)
print(ans)
```

```python
s=0
n=1
op=['a','b','c','d']
for k,v in q.items():
    print("q{}.{}".format(n,k))
    j=0
    for i in v:
        print("q{} {}".format(op[j],i))
        j=j+1
        d=input("give the correct answer:[a,b,c,d]")
        
        if d==ans[k]:
            s=s+1
    n=n+1
```
